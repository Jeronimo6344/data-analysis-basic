```python
from hossam import load_data
from matplotlib import pyplot as plt
import seaborn as sb
from matplotlib import font_manager as fm
import numpy as np
```

```python
my_dpi=200
font_path='../NotoSansKR-Regular.ttf'
fm.fontManager.addfont(font_path)
font_prop=fm.FontProperties(fname=font_path)
font_name=font_prop.get_name()
plt.rcParams['font.family']=font_name
plt.rcParams['font.size']=7
plt.rcParams['axes.unicode_minus']=False
```

```python
origin=load_data('penguins')
origin.head()
```

```python
# 1. 그래프 초기화
width_px=1280
height_px=760
rows=1
cols=1
figsize=(width_px/my_dpi, height_px/my_dpi)
fig, ax=plt.subplots(rows, cols, figsize=figsize, dpi=my_dpi)

# 2. 그래프 그리기
sb.barplot(data=origin, x='species', y='body_mass_g', hue='sex', estimator=np.mean, errorbar=('ci',98), palette='Set2')

ax.grid(True)
plt.tight_layout()
plt.show()
plt.close()
```

```python
# 1. 그래프 초기화
width_px=1280
height_px=760
rows=1
cols=1
figsize=(width_px/my_dpi, height_px/my_dpi)
fig, ax=plt.subplots(rows, cols, figsize=figsize, dpi=my_dpi)

# 2. 그래프 그리기
sb.countplot(data=origin, x='species', hue='sex', order=['Chinstrap','Gentoo','Adelie'],palette='Pastel1')

ax.grid(True)
plt.tight_layout()
plt.show()
plt.close()
```

```python
# 1. 그래프 초기화
width_px=1280
height_px=760
rows=1
cols=1
figsize=(width_px/my_dpi, height_px/my_dpi)
fig, ax=plt.subplots(rows, cols, figsize=figsize, dpi=my_dpi)

# 2. 그래프 그리기
sb.pointplot(data=origin, x='species', y='flipper_length_mm', hue='sex', estimator=np.mean, errorbar=('ci',95), linestyles='--', linewidth=1, markers='*' ,palette='Dark2')

ax.grid(True)
plt.tight_layout()
plt.show()
plt.close()
```
