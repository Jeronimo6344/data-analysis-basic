```python
from hossam import load_data
from matplotlib import font_manager as fm
from matplotlib import pyplot as plt
import seaborn as sb
```

```python
font_path='../NotoSansKR-Regular.ttf'
fm.fontManager.addfont(font_path)
font_prop=fm.FontProperties(fname=font_path)
font_name=font_prop.get_name()
font_name
```

```python
my_dpi=120
my_font_name='Noto Sans KR'
my_theme='dark'

sb.set_theme(style=my_theme)

plt.rcParams['font.family']=my_font_name
plt.rcParams['font.size']=16
plt.rcParams['axes.unicode_minus']=False
```

```python
#   1
width_px=1280
height_px=760
rows=1
cols=1
figsize=(width_px/my_dpi, height_px/my_dpi)
fig, ax=plt.subplots(rows, cols, figsize=figsize, dpi=my_dpi)

#  
sb.lineplot(x=[1,2,3,4],y=[98,82,73,92],
            c='#2548b1',linestyle=':',linewidth=5,
            marker='o',markersize=20,markerfacecolor='#00ff00',
            markeredgecolor='#ff0000',markeredgewidth=5)

#  
ax.set_title('    ', color='#0066ff', fontsize=22, fontweight=1000, pad=20)
ax.set_xlabel('', color='#ff6600', fontsize=16, labelpad=10)
ax.set_ylabel('', color='#ff00ff', fontsize=16, labelpad=10)
ax.set_xlim([0,6])
ax.set_ylim([70,100])
ax.set_xticks([1,2,3,4], ['1','2','3','4'])
ax.set_yticks(range(70,101,3),range(70,101,3))


#   2
plt.grid()
plt.tight_layout()
# plt.savefig('myplot.png',dpi=my_dpi)
plt.show()
plt.close()
```

```python
origin=load_data('traffic_acc')
origin
```

```python
df=origin.drop('',axis=1).groupby('').mean()
df
```

```python
width_px=1280
height_px=760
rows=1
cols=1
figsize=(width_px/my_dpi, height_px/my_dpi)
fig, ax=plt.subplots(rows, cols, figsize=figsize, dpi=my_dpi)

sb.lineplot(data=df, x=df.index, y='')

plt.grid()
plt.tight_layout()
# plt.savefig('myplot.png',dpi=my_dpi)
plt.show()
plt.close()
```
