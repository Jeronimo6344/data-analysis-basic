# 1-2. 동네 아이스크림 가게 매출 분석

icecream_sales 데이터셋을 사용하여 아래의 임무를 수행하여라.

1. 데이터의 전체적인 구조와 정보를 확인하고, 판매 수량에 결측치가 있는지 확인하시오.
    - 만약 결측치가 있다면, 해당 행을 제거하시오.
2. 판매 날짜를 인덱스로, 아이스크림 맛을 컬럼으로 하여 각 맛의 일별 판매량을 보여주는 피벗 테이블을 만드시오.
3. 위에서 만든 피벗 테이블을 다시 긴 형식(Long format)의 데이터로 분리하여라. 컬럼 이름은 원본 이름을 그대로 유지해야 한다.

```python
from hossam import load_data
from pandas import pivot_table, melt
```

```python
origin=load_data('icecream_sales')
origin.info()
origin.isna().sum()
```

```python
my_df=origin.dropna(subset=['Quantity'])
my_df
```

```python
my_pv=pivot_table(my_df,index='Date',columns='Flavor',values='Quantity')
my_pv
```

```python
my_pv2=my_pv.reset_index()
my_melt=melt(my_pv2, id_vars=['Date'],value_vars=['딸기', '민트초코', '바닐라', '초콜릿'],var_name='Flavor',value_name='Quantity').dropna().reset_index(drop=True)
my_melt
```

# 5. 직원 성과 분석을 위한 데이터 스케일링

'performance' 데이터 셋에 대해 다음 스케일링 기법들을 각 컬럼에 적용하시오.

1. 'MonthlySales' 컬럼에는 Min-Max 스케일링을 적용하여 0과 1 사이의 값으로 변환하시오.
2. 'CustomerSatisfaction' 컬럼에는 MaxAbs 스케일링을 적용하여 -1과 1 사이의 값으로 변환하시오.
3. 'ProjectHours'와 'InnovationScore' 컬럼에는 Standard 스케일링(표준화)을 적용하여 평균이 0, 표준편차가 1인 값으로 변환하시오.
4. 모든 스케일링 과정이 끝난 후, 'MonthlySales'(Min-Max 스케일링), 'CustomerSatisfaction'(MaxAbs 스케일링), 'ProjectHours'(Standard 스케일링), 'InnovationScore'(Standard 스케일링) 컬럼의 값을 모두 더했을 때, 합산 값이 가장 높은 직원의 'EmployeeID'를 찾아보시오.

```python
from hossam import load_data
import numpy as np
from sklearn.preprocessing import MinMaxScaler, MaxAbsScaler, StandardScaler
from pandas import DataFrame
```

```python
origin=load_data('performance')
origin
```

```python
x=origin[['MonthlySales']]
scaler=MinMaxScaler()
origin['MonthlySales_MinMax']=scaler.fit_transform(x)
origin.head()
```

```python
y=origin[['CustomerSatisfaction']]
scaler=MaxAbsScaler()
origin['CustomerSatisfaction_MaxAbs']=scaler.fit_transform(y)
origin.head()
```

```python
z=origin[['ProjectHours']]
scaler=StandardScaler()
origin['ProjectHours_Standard']=scaler.fit_transform(z)
origin.head()
```

```python
a=origin[['InnovationScore']]
scaler=StandardScaler()
origin['InnovationScore_Standard']=scaler.fit_transform(a)
origin.head()
```

```python
origin['Score']=origin['MonthlySales_MinMax']+origin['CustomerSatisfaction_MaxAbs']+origin['ProjectHours_Standard']+origin['InnovationScore_Standard']
origin.head()
```

```python
final_df=origin.sort_values(['Score'], ascending=False)
final_df.head(1).index[0]
```
