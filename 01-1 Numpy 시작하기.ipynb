# 라이브러리 참조

```python
import numpy as np
```

# 배열 생성

```python
arr=np.array([1,3,5,7,9])
print(type(arr))
arr
```

# 배열의 크기

```python
arr=np.array([10,30,50,70,90])
size=len(arr)
print('배열의 원소는 %d개 입니다.'%size)
```

# 배열의 인덱싱과 슬라이싱

```python
mydata=np.array([5,10,15,20,25])

print(mydata[0])
print(mydata[1])
print(mydata[3])

print(mydata[1:4])
print(mydata[-4:-1])
```

# 반복문을 통한 활용

```python
mydata=np.array([100,110,120,130,140])

for i,v in enumerate(mydata):
    print('%d번째 원소 >> %d'%(i,v))
```
