# 01-1 연습문제

Numpy의 ndarray를 활용하여
1. 정수 10, 20, 30, 40, 50으로 구성된 배열을 생성하고,
2. 배열의 길이(원소 개수)를 출력하며,
3. 배열의 첫 번째 원소와 마지막 두 개 원소를 각각 출력하고,
4. for문과 enumerate()를 이용해 인덱스와 값을 함께 출력하시오.

```python
import numpy as np
```

```python
myarray=np.array([10,20,30,40,50])

print('배열의 길이: %d'%len(myarray))
print('첫 번째 원소: %d'%myarray[0])
print('마지막 두 개 원소:',myarray[-2:])

for i, v in enumerate(myarray):
    print('%d번째 원소 >> %d'%(i,v))
```

# 01-2 연습문제

학생 6명의 점수 데이터가 다음과 같이 주어집니다.  
score=[72.5, 88.0, 91.5, 67.0, 80.0, 59.5]

1. 위 점수 리스트를 정수형 ndarray로 변환하시오.
2. 점수가 80점 이상인 학생의 점수만 추출하여 출력하시오.
3. 모든 학생에게 다음 기준으로 학점을 부여하고 결과를 출력하시오.
    - 90점 이상 -> 'A'
    - 80점 이상 -> 'B'
    - 70점 이상 -> 'C'
    - 그 외 -> 'F'

```python
import numpy as np
```

```python
score=[72.5, 88.0, 91.5, 67.0, 80.0, 59.5]

myscore=np.array(score,dtype='int')
print('원본 점수:',myscore)

print('80점 이상:',myscore[myscore>=80])

conditions=[(myscore>=90),
            (myscore>=80),
            (myscore>=70)]
value=['A','B','C']

print('학점 결과:',np.select(conditions,value,default='F'))
```

당신은 AI 교육기관에서 수강생들의 모의시험 데이터를 검수하는 일을 맡고 있습니다.  
시험 데이터는 '3회차 시험 * 4명 학생'의 점수로 구성되어야 합니다.  
다음 조건에 맞게 NumPy로 코드를 작성하세요.

1. 평균이 75, 표준편차가 5인 정규분포를 따르는 랜덤 점수 데이터를 3행 4열 배열로 생성하시오.
    - 단, 결과가 재현되도록 랜덤 시드를 100으로 고정해야 함.
2. 생성된 점수 중 80점 이상인 항목만 추출하시오.

# 01-4 연습문제

아래 조건에 따라 NumPy를 이용해 데이터 타입과 구조를 정리하세요.

1. 0부터 11까지의 정수를 가지는 배열을 생성하시오.
2. 이 배열을 실수형(float)으로 변환하시오.
3. 변환된 배열을 2행 6열 구조로 재배치하시오.
4. 이후, 열의 수에 관계없이 3행으로 자동 변환되도록 reshape() 함수를 이용하시오.
5. 마지막으로, 1차원 배열로 축소하여 결과를 출력하시오.

```python
import numpy as np
```

```python
arr1=np.arange(12)
print('원본 배열:',arr1)

arr2=arr1.astype('float')
print('실수형 변환:',arr2)

arr3=arr2.reshape(2,6)
print('2행 6열 형태:')
print(arr3)

arr4=arr3.reshape(3,-1)
print('자동 계산된 형태(3행, 4열):')
print(arr4)

arr5=arr4.reshape(-1)
print('1차원 배열로 축소:')
print(arr5)
```

'Alpha' 팀 점수 : [88,92,76,100,95,81,79,98,85,87]  
'Bravo' 팀 점수 : [65,99,78,72,80,91,75,82,89,70]

기초 통계량 -> 도수, 평균, 표준편차, 최소값, 1사분위수, 중위값, 3사분위수, 최대값

```python
import numpy as np
from pandas import DataFrame
```

```python
Alpha=[88,92,76,100,95,81,79,98,85,87]
Bravo=[65,99,78,72,80,91,75,82,89,70]

mydata={
    'alpha':{
        '도수':len(Alpha),
        '평균':np.average(Alpha),
        '표준편차':np.std(Alpha),
        '최소값':np.min(Alpha),
        '1사분위 수':np.quantile(Alpha,0.25),
        '중위값':np.median(Alpha),
        '3사분위 수':np.quantile(Alpha,0.75),
        '최대값':np.max(Alpha)
    },
    'bravo':{
        '도수':len(Bravo),
        '평균':np.average(Bravo),
        '표준편차':np.std(Bravo),
        '최소값':np.min(Bravo),
        '1사분위 수':np.quantile(Bravo,0.25),
        '중위값':np.median(Bravo),
        '3사분위 수':np.quantile(Bravo,0.75),
        '최대값':np.max(Bravo)
    }
}

df=DataFrame(mydata)
df
```

# 01-6 연습문제

주어진 데이터를 사용하여 판매량 분포를 파악하고, 아래 요구사항에 따라 분석 결과를 출력하는 Python 콛즈를 작성하세요.  
- 일일 판매량 데이터
    [18,25,33,21,28,40,15,3,9,17,22,29,38,45,11,6,31,24,27,35,12,19,39,42,7,2,20,34,26,30]
- 요구사항
    1. 주어진 일일 판매량 데이터를 'numpy'배열로 생성하세요.
    2. 판매량 구간을 '[0,10,20,30,40,50]'으로 직접 설정하여 도수분포를 분석하세요.
    3. 2번의 도수 분포표를 활용하여 절대도수, 상대도수, 누적도수를 포함하는 도수 분포표를 출력하세요.
        - 상대도수는 소수점 넷째 자리에서 반올림하여 셋째 자리까지 표시하세요.

```python
import numpy as np
from pandas import DataFrame
```

```python
mydata=np.array([18,25,33,21,28,40,15,3,9,17,22,29,38,45,11,6,31,24,27,35,12,19,39,42,7,2,20,34,26,30])
myarray=[0,10,20,30,40,50]
hist,bins=np.histogram(mydata,myarray)
count=0
# 도수분포표
s=len(hist)
for i in range(0,s):
    if i+1<s:
        tpl='%d이상 %d미만: 절대도수(%d), 상대도수(%0.3f), 누적도수(%d)'
    else:
        tpl='%d이상 %d이하: 절대도수(%d), 상대도수(%0.3f), 누적도수(%d)'
    print(tpl%(bins[i],bins[i+1],hist[i],(hist[i]/np.sum(hist)),count+hist[i]))
    count+=hist[i]
```
