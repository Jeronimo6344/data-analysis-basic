# 배열의 연산

```python
arr1=np.array([10,15,20,25,30])
arr2=np.array([2,3,4,5,6])
print(arr1)
print(arr2)
```

```python
a1=arr1+arr2
print(a1)

a2=np.add(arr1,arr2)
print(a2)
```

```python
b1=arr1-arr2
print(b1)

b2=np.subtract(arr1,arr2)
print(b2)
```

```python
c1=arr1*arr2
print(c1)

c2=np.multiply(arr1,arr2)
print(c2)
```

```python
d1=arr1/arr2
print(d1)

d2=np.divide(arr1,arr2)
print(d2)
```

# 배열의 조건식

```python
grade=np.array([82,77,91,88])
grade
```

```python
bool_array=np.array([True,False,True,False])
bool_array
```

```python
result1=grade[bool_array]
result1
```

```python
mypoint=np.array([82,77,91,88])
ge80=mypoint[mypoint>=80]
ge80
```

```python
ge80le90=mypoint[np.logical_and(mypoint>=80,mypoint<=90)]
ge80le90
```

```python
gt_or_lt=mypoint[np.logical_or(mypoint<80,mypoint>90)]
gt_or_lt
```

```python
exam=np.array([60,83,75,92,86])
result=np.where(exam>=80,'합격','불합격')
result
```

```python
exam=np.array([60,85,51,79,94,88])

conditions=[(exam>=90),
            (exam>=80),
            (exam>=70)]
value=['A','B','C']

result=np.select(conditions,value,default='F')
result
```
