```python
import numpy as np
from IPython.display import Image
from wordcloud import WordCloud
from wordcloud import STOPWORDS
from PIL import Image as Img
```

```python
with open('wordcloud/이상한나라의앨리스.txt', 'r', encoding='utf-8') as f:
    text=f.read()
```

```python
wc=WordCloud(width=1200, height=800)
wc.generate(text)
wc.words_
```

```python
wc.to_file('simple.png')
```

```python
Image('simple.png')
```

```python
ignore=set(STOPWORDS)
ignore.add('said')
ignore.add('Alice')
```

```python
color='#FFFFFF'
```

```python
img=Img.open('wordcloud/앨리스배경.png')
iarr=np.array(img)
iarr
```

```python
wc=WordCloud(width=600, height=1200, max_font_size=150, scale=2.0, max_words=100, stopwords=ignore, mask=iarr)
wc.generate(text)
wc.to_file('options.png')
Image(filename='options.png')
```

```python
wc=WordCloud(width=1200, height=800, stopwords=ignore, background_color='#FFFFFF', colormap='ocean')
wc.generate(text)
wc.to_file('colormap.png')
Image(filename='colormap.png')
```
