```python
import numpy as np
```

```python
myarr=np.array([98,72,80,64,88,65,79,59,81,74])
myarr
```

```python
print('총 도수:',len(myarr))
```

```python
print('최대값:',np.max(myarr))
print('최소값:',np.min(myarr))
print('합계:',np.sum(myarr))
```

```python
print('평균:',np.average(myarr))
```

```python
size=len(myarr)
sorted_arr=np.sort(myarr)
print('정렬:',sorted_arr)

if size%2==0:
    p=(size//2)-1
    q=p+1
    median=(sorted_arr[p]+sorted_arr[q])/2
else:
    p=size//2
    median=sorted_arr[p]

print('중앙값:',median)
```

```python
print('중앙값:',np.median(myarr))
```

```python
x=np.sort(myarr)
n=len(x)
q1_pos=(n-1)*0.25
i=int(np.floor(q1_pos))
f=q1_pos-i

Q1=x[i]+f*(x[i+1]-x[i])
print('직접 계산한 1사분위수 (Q1):',Q1)
print('Numpy로 계산한 1사분위수:',np.quantile(myarr,0.25))
```

```python
x=np.sort(myarr)
n=len(x)
q3_pos=(n-1)*0.75
i=int(np.floor(q3_pos))
f=q3_pos-i

Q3=x[i]+f*(x[i+1]-x[i])
print('직접 계산한 3사분위수 (Q3):',Q3)
print('Numpy로 계산한 3사분위수:',np.quantile(myarr,0.75))
```

```python
print('데이터 구간:', np.max(myarr)-np.min(myarr))
```

```python
print('사분위수 구간(IQR):',np.quantile(myarr,0.75)-np.quantile(myarr,0.25))
```

```python
print('분산:',np.var(myarr))
print('표준편차:',np.std(myarr))
```
